## cargar base de datos SQL con Pandas

```python
path = 'database.sqlite'
sqliteConnection = sqlite3.connect(path) #conexiÃ³n a la base de datos sql
```

creo un query para filtrar y cargar los datos.

```python
query = 'SELECT * FROM Tweets LIMIT 10'
pd.read_sql_query(query,sqliteConnection) # lectura de base de datos, con el query
```

### Select, From y Where

SELECT pull specific columns from a table,
WHERE pull rows that meet specified criteria

```python
query = "SELECT negativereason, negativereason_confidence \
        FROM Tweets \
        WHERE airline_sentiment='negative' \
        LIMIT 10"
pd.read_sql_query(query,sqliteConnection)
```

### Group by, Having, COUNT/AVG/MIN/MAX, Order by

GROUP BY takes the name of one or more columns, and treats all rows with the same value in that column as a single group when you apply aggregate functions like COUNT().

HAVING is used in combination with GROUP BY to ignore groups that don't meet certain criteria.

ORDER BY sorts the results returned by the rest of your query. You can reverse the order using the DESC argument.

```python
query = "SELECT negativereason,AVG(negativereason_confidence)\
        FROM Tweets\
        GROUP BY negativereason \
        HAVING airline_sentiment='negative'\
        ORDER BY AVG(negativereason_confidence)\
        "
pd.read_sql_query(query,sqliteConnection)
```
